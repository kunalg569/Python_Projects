## Addressing outliers:


```python
# List of columns for which you want to generate box plots
cols = ['adr', 'days_in_waiting_list']

# Create subplots
fig, axs = plt.subplots(1, len(cols), figsize=(15, 5))

# Iterate through columns and generate box plots
for i, col in enumerate(cols):
    df.boxplot(column=col, ax=axs[i])
    axs[i].set_title(col)
    
print('We have 1 outlier in adr and several in days_in_waiting_list ')
```

```python
df[df['adr'] > 5000]
```

```python
# Removing adr outlier
df.drop(index=48515, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
# Removing the days_in_waiting_list outlier:

# calculate 25th percentile of annual strikes:: Q1
percentile25 = df['days_in_waiting_list'].quantile(0.25)

# calculate 75th percentile of annual strikes:: Q3
percentile75 = df['days_in_waiting_list'].quantile(0.75)

# calculate interquartile range
iqr = percentile75-percentile25

# upper limit:
upper_limit = percentile75 + 1.5 * iqr
# lower limit
lower_limit = percentile25 - 1.5 * iqr
```

```python
print(upper_limit)
```

Since we have upper limit as 0 it means we have max people booking without waiting. It is best to leave the outliers untouched as realistically there can be properties that have several days of waiting period. For sake of integrity of the dataset we will drop anything above 250 waiting days since our plots show most of the values lie beneath it.

```python
# List of columns for which you want to generate box plots
cols = ['adr', 'days_in_waiting_list']

# Create subplots
fig, axs = plt.subplots(1, len(cols), figsize=(15, 5))
fig.suptitle('Outliers Removed', size = 20)

# Iterate through columns and generate box plots
for i, col in enumerate(cols):
    df.boxplot(column=col, ax=axs[i])
    axs[i].set_title(col)
    
print('We have 1 outlier in adr and several in days_in_waiting_list ')
```

```python
# Filtering only realistic values
df = df[df['days_in_waiting_list'] < 250]

df = df[df['adr'] < 5000]
```

## Reservation status in different hotels

```python
plt.figure(figsize=(6, 4))
ax1 = sns.countplot(data=df, x='hotel', hue='is_canceled', palette='Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(legend_labels, ['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Reservation status in different hotels')
plt.xlabel('Hotel Type')
plt.ylabel('Count')
plt.show()

total_cancellations_city = df[(df['is_canceled'] == 1) & (df['hotel'] == 'City Hotel')].shape[0]
total_cancellations_resort = df[(df['is_canceled'] == 1) & (df['hotel'] == 'Resort Hotel')].shape[0]

total_count_city = df[df['hotel'] == 'City Hotel'].shape[0]
total_count_resort = df[df['hotel'] == 'Resort Hotel'].shape[0]

cancel_city_ratio = total_cancellations_city / total_count_city
cancel_resort_ratio = total_cancellations_resort / total_count_resort

print('City Cancellation ratio:', cancel_city_ratio)
print('Resort Cancellation ratio:', cancel_resort_ratio)
```

**Observation**:
- Cancelaltion ratio is almost twice for City hotel than Resort hotel.

## Does price have any effect on the cancellations

```python
# Avg rate of properties typically

# Filter out resrot and city hotel entires
resort_hotel = df[df['hotel'] == 'Resort Hotel']
city_hotel = df[df['hotel'] == 'City Hotel']

# find mean pricing based on reservation date 
resort_hotel_prices = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
# resort_hotel_prices = resort_hotel_prices.reset_index()

city_hotel_prices = city_hotel.groupby('reservation_status_date')[['adr']].mean()
#city_hotel_prices = city_hotel_prices.reset_index()

# Plotting by common dates
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', size = 20)
plt.plot(resort_hotel_prices.index, resort_hotel_prices['adr'], label = 'Resort Hotel')
plt.plot(city_hotel_prices.index, city_hotel_prices['adr'], label = 'City Hotel')
plt.legend(fontsize =15)
plt.show()
```

**Obeservation**:
The average pricing for city hotels is comparatively less than resort hotel

## Reservation distribution based on months:

```python
# Add column 'month'
df['month'] = df['reservation_status_date'].dt.month

# Visualize

plt.figure(figsize = (15,8))
ax = sns.countplot(data= df, x = 'month', hue='is_canceled', palette='Blues')

legend_labels = ax.get_legend_handles_labels()
ax.legend(legend_labels, ['Not Canceled', 'Canceled'], bbox_to_anchor = (1,1))

plt.title('Reservation Count per Month', size = 20)
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(['not canceled', 'canceled'], fontsize = 12)
plt.show()
```

**Observation**:
January recorded the highest number of cancellations, surpassing its reservation count for the month.

## Average pricing per month during cancellations:

```python
# Grouping by month
avg_month_price = df[df['is_canceled'] == 1].groupby('month')[['adr']].mean()

plt.figure(figsize = (15, 5))
sns.barplot(avg_month_price.index, avg_month_price['adr'], palette = 'Blues')
plt.title('Mean pricing of hotels during cancellations', size = 20)
```

```python
# Avg pricing distribution for cancellations and non cancellations 

cancel = df[df['is_canceled'] == 1].groupby('reservation_status_date')[['adr']].mean()
cancel = cancel.reset_index()
cancel = cancel.sort_values('reservation_status_date', ascending=True)

no_cancel = df[df['is_canceled'] == 0].groupby('reservation_status_date')[['adr']].mean()
no_cancel = no_cancel.reset_index()
no_cancel = no_cancel.sort_values('reservation_status_date', ascending=True)

plt.figure(figsize =(20,6))
plt.title('Average Daily Rate', size = 20)
plt.plot(cancel['reservation_status_date'], cancel['adr'], label = 'Canceled')
plt.plot(no_cancel['reservation_status_date'], no_cancel['adr'], label = 'Not Canceled')
```

```python
# Shrinking out data since we do not have enough data for 2015 and less than 2017-09
cancel_df = cancel[(cancel['reservation_status_date'] > '2016') & (cancel['reservation_status_date'] < '2017-09')]
no_cancel_df = no_cancel[(no_cancel['reservation_status_date'] > '2016') & (no_cancel['reservation_status_date'] < '2017-09')]

plt.figure(figsize =(20,6))
plt.title('Average Daily Rate', size = 20)
plt.plot(cancel_df['reservation_status_date'], cancel_df['adr'], label = 'Canceled')
plt.plot(no_cancel_df['reservation_status_date'], no_cancel_df['adr'], label = 'Not Canceled')
plt.legend(fontsize = 15)
```

**Observation**: <br>
Cancellations exclusively hinge on pricing, as evidenced by various instances. **Our hypothesis is true**
