## Buidling MLR model 1: 
(Based on weather conditions only)

```python
ols_formula = 'cnt ~ C(weathersit) + temp + hum + windspeed'

OLS = ols(formula=ols_formula, data=weather_final)

model_1 = OLS.fit()

model_1_results = model_1.summary()
```

```python
model_1_results
```

## Checking model assumptions:
- Linearity assumption
- Independant observations
- Residuals normality assumption
- Homoscedasticity (constant variance in residuals VS fitted_values)
- No multicollinearity 

### Linearity Assumption:
- All the independant variables have a linear relationship with dependant variable

```python
sns.pairplot(weather_final)
```

**Observation**:
- There seems to be a linear relationship only in case temp and cnt
- Conclusion, our linearity assumption is $NOT$ met

### Independant Observations Assumption:
- The observations in our data should be independant of one another
- In our case, all the observations recorded are independant of each hence our assumption is true

### Normality assumption Assumption:
- The residuals of the model have a normal distribution and the qqplot has to be diagonal straight line

```python
fig, ax = plt.subplots(1,2,figsize = (10,5))

sns.histplot(model_1.resid, ax=ax[0])
ax[0].set_xlabel('Residual Values')
ax[0].set_title('Histogram of residuals')

sm.qqplot(model_1.resid, line = 's', ax=ax[1])
ax[1].set_title('Normal QQ plot')
```

**Oberservaiton:**
Our normality assumption is met

### Homoscedasticity Assumption:
- Our residuals must have constant variance, in simple words, our fitted values VS residuals plot should not have any pattern

```python
sns.scatterplot(x=model_1.fittedvalues, y=model_1.resid)
```

## No multicollinearity assumption:
The **no multicollinearity assumption** states that no two independent variables ($X_i$ and $X_j$) can be highly correlated with each other. 

Two common ways to check for multicollinearity are to:

* Create scatterplots to show the relationship between pairs of independent variables
* Use the variance inflation factor to detect multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = weather[['weathersit', 'temp', 'hum', 'windspeed']]

# Calculate VIF for each variable
vif = pd.DataFrame()
vif["Variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
```

```python
vif
```

**Observations**:
- The variables seem to have high multicollineairy 

**Conclusion:**
- It would be beneficial if all the independant variables are considered for the analysis as the features highlighting weather conditions not only have strong multicollineairty but also explain a mere 48% varaince on the our target varaible 'cnt which is our demand.
- In other words our model has a bias or in other words, underfits the data

#### Check for mean demand per categorical feature using groupby()


```python
columns = ['season', 'mnth', 'weathersit', 'holiday', 'weekday','workingday']
for i in columns:
    print(df2.groupby(i)['cnt'].mean())
```

### **Observation:**

- The *categorical means* appear to be evenly distributed across variables, suggesting **no significant influence** on the target variable.
- **Exceptions:** Some uneven distribution of demand across the twelve months and lower demand observed during weather situations categorized as "*Light snow, light rain with thunderstorms, and scattered clouds*" (coded as 3 in the *weathersit* variable).


- Given the *absence of concrete evidence* to justify the selection of specific variables for our Multiple Linear Regression (MLR), we opt to employ **Recursive Feature Elimination (RFE)** from the **SciKit Learn** library.
  
  - RFE is chosen for its compatibility and utility with **sklearn**, enabling a systematic approach to feature selection.
  - RFE, or Recursive Feature Elimination, is a feature selection technique used in machine learning. It works by recursively removing attributes and building a model on those attributes that remain.
  - It assesses the importance of each feature and eliminates the least important ones, iteratively. This process helps to identify the most relevant features for improving model performance.
  
**Future Steps:**
- We will first rescacle our continuous features using MinMaxSacler then split our data as train and test
- Then, use our Linear Reg model along with RFE to select the most relevant features
- Once, we have the relevant features only, we will check their variation inflation factor for any strong correlation between our independant variables as it violates one of the MLR model assumption; features are independant from each other
- Then, add a constant to the features selected which is nothing but the *intercept*. We thus allow the model to predict this constant during the fitting process. 
    - *Note*:
        - If you don't add the constant term to the feature matrix, the linear regression model will still try to fit a line to the data, but it will be forced to go through the origin (0,0).<br> However, this assumption of the model might not be appropriate for all datasets. 
        - For instance, if the relationship between the response variable and the predictor variables does not necessarily pass through the origin, omitting the constant term can introduce bias in the estimation of coefficients and affect the overall performance of the model.

- We will then eliminate any features having strong correlation and build the model.
- This is iterative process till we have the appropriate features

```python
# Creating our dummy variables for categorical variables
bike = pd.get_dummies(df2, drop_first=True)
bike.info()
```

**Rescaling of features**:

```python
# Initializing scaler values:
scaler = MinMaxScaler()
```

```python
num_vars = ['temp', 'atemp', 'hum', 'windspeed','cnt']

# rescaling the continuous features
bike[num_vars] = scaler.fit_transform(bike[num_vars])

bike.describe()
```

```python
# Splitting our data
X = bike.drop('cnt', axis = 1)
y = bike[['cnt']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Running a RFE with output number of features = 15:
lm = LinearRegression()
lm.fit(X_train, y_train)

# Running RFE
rfe = RFE(estimator=lm, n_features_to_select=15)  
rfe = rfe.fit(X_train, y_train)

list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Features selected by RFE
col = X_train.columns[rfe.support_]
col
```

```python
# Features NOT selected by RFE
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# VIF score

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()

vif['Features'] = X_train_rfe.columns

vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]

vif['VIF'] = round(vif['VIF'], 2)

vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Add a constant
X_train_lm1 = sm.add_constant(X_train_rfe)
```

```python
# Create a first fitted model
lr1 = sm.OLS(y_train, X_train_lm1).fit()
```

```python
# Print a summary of the linear regression model obtained
print(lr1.summary())
```
