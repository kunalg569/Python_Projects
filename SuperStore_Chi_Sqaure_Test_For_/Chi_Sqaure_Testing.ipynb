## Chi Sqaured Test: 

The Chi-Square test is a statistical procedure for determining the difference between observed
and expected data. This test can also be used to determine whether it correlates to the
categorical variables in our data. It helps to find out whether a difference between two
categorical variables is due to chance or a relationship between them.

The Chi-Square test statistic is calculated as:

$$\chi^2_c = \sum \left( \frac{(O_i - E_i)^2}{E_i} \right)$$

Where:
- \( c \) = Degrees of freedom
- \( O \) = Observed Value
- \( E \) = Expected Value


````python
# Segregating our categorical variables in question:
contingency_table = pd.crosstab(df['Category'], df['Ship Mode'])
contingency_table
````

## Steps for Chi Sqaure test for independence:

- Step 1: What you have to achieve? Prove (or disprove) that there is dependance between the Shipping mode and Product Category
- Step 2: State Hypothesis, $H_0$ and $H_a$
    - $H_0$: There is dependancy between the Ship Mode and Product Category
    - $H_1$: There is **no** dependancy between the Ship Mode and Product Category
- Step 3: Decide Significance level, ie: alpha = 5%. If p-value is less than 0.05 we reject the Null Hypothesis
- Step 4: Calculate Chi-Sqaure statitics using **stats.contingency.chi2_contingency()**
- Step 5: Calculate p-value
- Step 6: Accept or reject $H_0$

````python
# Fetching observed values:
observed_values = contingency_table.values
observed_values
````

````python
# Calculating our expected Values:
chi2_stat, p_val, dof, expected_values = stats.chi2_contingency(contingency_table)

expected_values
````

````python
# Calculate Chi sqaured stat:
result = stats.contingency.chi2_contingency(observed_values, correction=False)
````

````python
result
````

### Chi-Square Test Results for Independance

After performing the Chi-Square Test of Independence on our contingency table, we have obtained the following results:

- **Chi-squared statistic**: `3.7418927103679342`
  - This is the test statistic that tells us how much the observed frequencies differ from the expected frequencies.

- **p-value**: `0.7115573960981196`
  - This is the probability of observing the data, or something more extreme, if the null hypothesis is true. Since this value is much greater than `0.05`, we fail to reject the null hypothesis, implying there is no significant association between the variables.

- **Degrees of freedom**: `6`
  - This value is based on the number of categories minus one in each dimension of the contingency table.

- **Expected frequencies**:
  ```python
  [
   [ 318.3242415 ,  114.09623046,  403.15415262, 1240.42537542],
   [ 905.13872714,  324.4268056 , 1146.34824803, 3527.08621923],
   [ 277.53703136,   99.47696394,  351.49759935, 1081.48840535]
  ]

*Since our p-value(0.711) > 0.05, we fail to reject our null hypothesis. This means there no statistical evidence of dependency between the Shipping Mode and the Product Category*

## Steps for Chi Sqaure test for goodness of fit:

- Step 1: What you have to achieve? Prove (or disprove) that there is dependance between the Shipping mode and Product Category
- Step 2: State Hypothesis, $H_0$ and $H_a$
    - $H_0$: The product category based off of the Shipping mode follows an expected distribution
    - $H_1$: The product category based off of the Shipping mode does not follow an expected distribution
- Step 3: Decide Significance level, ie: alpha = 5%. If p-value is less than 0.05 we reject the Null Hypothesis
- Step 4: Calculate Chi-Sqaure statitics using stats.contingency.chi2_contingency()
- Step 5: Calculate p-value
- Step 6: Accept or reject $H_0$

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure your DataFrame is named 'df', and the column names are 'Category' and 'Ship Mode'
sns.countplot(x='Ship Mode', hue='Category', data=df)

plt.title('Count of Categories for each Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.show()
````

````python
result_goodness = stats.chisquare(f_obs=list(observed_values), f_exp=list(expected_values))

result_goodness
````

````python
# Extract the maximum test statistic (which is equivalent to the discrepancy)
max_statistic = max(result_goodness.statistic)

# Calculate a single p-value based on the maximum test statistic
single_p_value = stats.chi2.sf(max_statistic, df=len(observed_values)-1)

print("Single p-value:", single_p_value)
````

*Since our p-value(0.4127) > significance level(0.05) we fail to reject our null hypothesis. This means the product category based off of shipping mode do not follow an expected distribution.*
