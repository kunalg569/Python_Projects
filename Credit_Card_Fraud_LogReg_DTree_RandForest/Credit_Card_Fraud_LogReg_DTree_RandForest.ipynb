```python
# Data Manipulation imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Pre precosseing import
from sklearn.preprocessing import RobustScaler

# Machine learning imports
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Evaluation matrics imports
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve, auc

# Setting display size
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# ignoring warnings
import warnings
warnings.filterwarnings('ignore')
```

**Observation**: 
- We dont really have anything insightful since the values are trasnformed using PCA. 
- One peculiarity¬†of this dataset is the inclusion of outliers in our analysis. This is becasue an outlier might be the case of a fraudulent transaction. We would want to preserve this to prevent loss of information from our dataset.
- One thing to note is the values in the `Amount`column; these are on a much differnet scale than the rest of the values. We will transform these values using **Robust Scaler**. Since most of the values are concentrated around zero having a few outliers as the plots range shows, we need to inlcude these values in our analysis for potential frauduleant transaction.
- With respect to the time scale, since the distribution is even, we will normalize these values to have them from range 0-1.

> By not rescaling all values to the range of 0-1, you're preserving the outliers in the dataset, which is important for detecting potentially fraudulent transactions. Rescaling everything to a common scale could lead to loss of information, particularly regarding outliers, which may be critical for fraud detection.


```python
# Plot histograms for data 
df.hist(bins=30, figsize = (30,30))
```

```python
# Changing 'Amount' scale
new_df = df.copy()

new_df['Amount'] = RobustScaler().fit_transform(new_df['Amount'].to_numpy().reshape(-1,1))

new_df['Amount'].hist()
print(new_df['Amount'].describe())
```

**Note**: 
- The rescaled values are not within the range of 0-1 or -1 to 1. Though this is the case, the values have been rescaled from the range of 0-25000 to 0-350. 
- Moreover, the std dev has been reduced from 250 to 3. This change will prove benficial for our modelling phase as the values are more aligned with each other. 
- Furthermore, there are outliers in the dataset as the rescaled values suggest but the potential bias seems to have been reduced as the dsitribution remains the same. This will help our models learn better on our dataset.

```python
# Changing time scale
time = new_df['Time']
new_df['Time'] = (time - time.min()) / (time.max() - time.min())
```

```python
# Overview of change
new_df['Time'].describe()
```

```python
# Checking the distribution once again

new_df.hist(bins = 30, figsize = (30,30))
```

## We will now address the imbalance in the dataset for the target class

```python
# Distribution:
new_df['Class'].value_counts()
```

We will perform **Downsampling** to match the 492 number of fraudulant transactions 

```python
from sklearn.utils import resample

# Assume df is your dataframe and 'verified_status' is the column with class labels

# Separating the majority and minority classes
data_minority = df[df['Class'] == 1]
data_majority = df[df['Class'] == 0]

# Calculate the number of samples needed from the majority class to achieve 50-50 split
# Let's say we want the majority class to be 60% of the dataset

majority_size = int(len(data_minority) * (50/50)) # note: 60/40 = desired_majority_proportion / desired_minority_proportion

# Downsample majority class to the calculated size
data_downsample_majority = resample(data_majority, 
                                    replace=False,  # sample without replacement
                                    n_samples=majority_size,  # calculated size for desired distribution
                                    random_state=0)  # reproducible results

# Combine minority class with downsampled majority class
data_downsampled = pd.concat([data_downsample_majority, data_minority])

# Resetting the index
data_downsampled = data_downsampled.reset_index(drop=True)

# Displaying the first few rows of the downsampled dataframe
print(data_downsampled['Class'].value_counts())
```

```python
# Making a copy
new_df = data_downsampled.copy()

new_df.info()
```

## Modelling:

## Logistic Regression

Moving on to the modelling phase. We will be using the following models:

- **Logistic Regression**: Logistic Regression is a commonly used algorithm for binary classification tasks like fraud detection. However, it's essential to note that Logistic Regression assumes a linear relationship between the features and the log-odds of the target variable, which may not always hold true. Nevertheless, it's a good starting point for classification tasks.
- **Decision Trees with Cross-Validation**: Decision Trees are intuitive and easy to interpret, making them suitable for understanding the underlying patterns in the data. By using cross-validation, you're effectively evaluating the model's performance on multiple subsets of the data, which helps in estimating its generalization ability.
- **Random Forest with Cross-Validation using Grid Search CV**: Random Forest is an ensemble learning method that combines multiple decision trees to improve predictive performance and reduce overfitting. By using cross-validated Grid Search CV, you're tuning the hyperparameters of the Random Forest model to find the best combination that maximizes performance.


```python
# Since we concatenated datasets[majority and minority classes] we will shuffle the dataset rows for randomization
df_shuffled = new_df.sample(frac=1).reset_index(drop=True)

# Custom split to have equal testing and validation data points
X_train, y_train = df_shuffled.iloc[:700, :-1].values, df_shuffled.iloc[:700, -1].values.astype(int)
X_test, y_test = df_shuffled.iloc[700:842, :-1].values, df_shuffled.iloc[700:842, -1].values.astype(int)
X_val, y_val = df_shuffled.iloc[842:, :-1].values, df_shuffled.iloc[842:, -1].values.astype(int)

# Size Check
X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
# Instantiate a model object
log_reg = LogisticRegression()

# Hyperparameter Tuning
param_grid = {
    'C': [0.1, 1, 10, 100], 
    'penalty': ['l1', 'l2'], 
    'solver': ['liblinear']
}

# Grid Search for best logistic regression model:
log_cv = GridSearchCV(log_reg, param_grid, cv=5, scoring='f1', verbose=1)
```

```python
# Fit the model
%%time
log_cv.fit(X_train, y_train)
```

```python
# Best LogReg model:
log_reg_cv = log_cv.best_estimator_
```

```python
# Get the predicted probabilities for the positive class (class 1) on validation set
y_val_pred_prob = log_reg_cv.predict_proba(X_val)[:, 1]

# Compute ROC curve and ROC area for validation set
fpr_val, tpr_val, _ = roc_curve(y_val, y_val_pred_prob)
roc_auc_val = auc(fpr_val, tpr_val)

# Plot ROC curve for validation set
plt.figure()
lw = 2
plt.plot(fpr_val, tpr_val, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc_val)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - Validation Set')
plt.legend(loc="lower right")
plt.show()
```

- **ROC curve**: The orange line represents the ROC curve, which shows the trade-off between the TPR and FPR at various threshold settings. The better the model is at distinguishing between the positive and negative classes, the closer the curve will be to the upper left corner of the graph.
- **Area Under the Curve (AUC)**: The AUC value is a measure of the model‚Äôs overall ability to discriminate between positive and negative classes. An AUC of 0.91, as shown in the graph, is quite high, indicating that the model has good discriminatory ability.
- **Dotted Line**: The blue dotted line represents a random classifier (AUC = 0.5), where the true positives and false positives are equal across all thresholds. The further the ROC curve is above this line, the better the model performs compared to a random guess.

```python
y_pred_log_reg = log_reg_cv.predict(X_val)
print(classification_report(y_val, y_pred_log_reg, target_names=['Not Fraud', 'Fraud']))
```


The classification report indicates that the model performs well in identifying fraudulent transactions, with high precision (0.90) and recall (0.81). However, it also maintains good balance in identifying non-fraudulent transactions, with a precision of 0.79 and recall of 0.89. Both classes have decent F1-scores: 0.83 for "Not Fraud" and 0.86 for "Fraud.

```python
# Function for plotting confusion matrix
def conf_matrix_plot(model, x_data, y_data):
    '''
    Accepts as argument model object, X data (test or validate), and y data (test or validate). 
    Returns a plot of confusion matrix for predictions on y data.
    ''' 
  
    model_pred = model.predict(x_data)
    cm = confusion_matrix(y_data, model_pred, labels=model.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=model.classes_)
  
    disp.plot()
    plt.title('Confusion Matrix for Credit Card Transactions', size = 12)
    plt.show()
```

```python
# Plotting Confusion matrix using function
conf_matrix_plot(log_reg_cv, X_val, y_val)
```

- Model has a particularly high precision for detecting fraudulent transactions, which is crucial for a fraud detection system to minimize the number of false alerts(6) that would be costly and inconvenient for customers.
- The recall for fraudulent transactions is also high, meaning it's capable of catching a good portion of fraudulent activities. However, there's room for improvement since false negative(6) rate for fraud means some fraudulent transactions are still being missed.
- The F1-score is quite balanced, especially for the "Fraud" class, which suggests that the model has achieved a good balance between precision and recall for this class.

```python
# Evalution metrics
accuracy = accuracy_score(y_val, y_pred_log_reg)
precision = precision_score(y_val, y_pred_log_reg)
recall = recall_score(y_val, y_pred_log_reg)
f1 = f1_score(y_val, y_pred_log_reg)

# Create a DataFrame to store the results
results_log_reg = pd.DataFrame({
    'Model': ['Logistic Regression CV'],
    'Accuracy': [accuracy],
    'Precision': [precision],
    'Recall': [recall],
    'F1 Score': [f1]
})
```

```python
results_log_reg_df = results_log_reg.copy()
results_log_reg_df
```

## Decision Tree using Cross validation

```python
# Instantiate the classifier
tuned_decision_tree = DecisionTreeClassifier(random_state = 42)

# Assign a dictionary of hyper parameters to search over
tree_para = {'max_depth':[4,5,6,7,8,9,10],
             'min_samples_leaf': [2, 5, 10, 20, 50],
             'min_samples_split': [2, 4, 6, 8]}

# Assign a dictionary of scoring metrics to capture
scoring = {'accuracy', 'precision', 'recall', 'f1'}

# Instantiate the GridSearch
clf = GridSearchCV(tuned_decision_tree, # The classifier (tuned_decision_tree)
                   tree_para,           # dictionary of hyper parameters to search over
                   scoring = scoring,   # dictionary of scoring metrics (scoring)
                   cv=5,                # cross-validation folds you want
                   refit="f1")          # The scoring metric that you want GridSearch to use when it selects the "best" model (i.e., the model that performs best on average over all 												   validation folds) (refit='f1'*)
```

```python
%%time
# Fit the model
clf.fit(X_train, y_train)
```

```python
# Best estimator 
dtree_cv_best = clf.best_estimator_
dtree_cv_best
```

```python
# Get the predicted probabilities for the positive class (class 1) on validation set
y_val_pred_prob = dtree_cv_best.predict_proba(X_val)[:, 1]

# Compute ROC curve and ROC area for validation set
fpr_val, tpr_val, _ = roc_curve(y_val, y_val_pred_prob)
roc_auc_val = auc(fpr_val, tpr_val)

# Plot ROC curve for validation set
plt.figure()
lw = 2
plt.plot(fpr_val, tpr_val, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc_val)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - Validation Set')
plt.legend(loc="lower right")
plt.show()
```

**Observations**
- The AUC of 0.91 suggests the model has a high ability to distinguish between classes.
- The ROC curve stays well above the diagonal line of no-discrimination, confirming good model sensitivity.
- The closer the curve follows the left-hand border and then the top border of the ROC space, the more accurate the test.

```python
# Classificaiton report
y_pred_dtree = dtree_cv_best.predict(X_val)
print(classification_report(y_val, y_pred_dtree, target_names=['Not Fraud', 'Fraud']))
```

**Observations**
- Moderate precision for fraud detection at 0.87: the model is accurate when it predicts fraud.
- Moderate recall for non-fraud at 0.88: it correctly identifies most non-fraudulent transactions.
- The F1-score is balanced for both classes, slightly favoring no fraud detection at 0.91.
- The overall accuracy of the model is 0.90, which is the proportion of total transactions correctly classified.
- There is still room for improvement since we have imbalances in the predictions. We will use a Random forest model further for the improvement
